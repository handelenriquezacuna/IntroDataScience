# Ejercicio: Análisis de Datos de Ventas y Publicidad
---



Contexto:
Tienes los datos de ventas de un producto durante 12 semanas, así como los gastos en publicidad durante el mismo período. Queremos analizar la relación entre gastos en publicidad y ventas.

El objetivo es entender cómo varían las ventas y los gastos, y ver si existe una relación entre ambos.

Datos:
Ventas (en unidades vendidas) durante 12 semanas:
[150, 170, 180, 210, 190, 230, 250, 280, 320, 300, 310, 350]

Gastos en publicidad (en miles de dólares) durante 12 semanas:
[10, 12, 14, 18, 16, 20, 22, 24, 28, 26, 30, 32]

Pistas:
Para la correlación, recuerda que un valor positivo indica una relación directa (a medida que aumenta una variable, también aumenta la otra).
Para las visualizaciones, asegúrate de que los gráficos sean claros y etiquetados correctamente para que los resultados sean fácilmente interpretables.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Tareas a Resolver:
# Estadísticas Descriptivas:
ventasUnidades = [150, 170, 180, 210, 190, 230, 250, 280, 320, 300, 310, 350]
gastosPublicidad = [10, 12, 14, 18, 16, 20, 22, 24, 28, 26, 30, 32]
df = pd.DataFrame({'Ventas': ventasUnidades, 'Gastos en Publicidad': gastosPublicidad})
correlacion = df['Ventas'].corr(df['Gastos en Publicidad'])
print("Correlación entre ventas y gastos en publicidad:", correlacion)
# Calcula la media, mediana, moda, rango y desviación estándar tanto para ventas como para gastos en publicidad.
print("Estadistica Descriptiva de Ventas")
ventasMedia = np.mean(ventasUnidades)
print("Media de ventas:", ventasMedia)
ventasMediana = np.median(ventasUnidades)
print("Mediana de ventas:", ventasMediana)
ventasModa = stats.mode(ventasUnidades)
print("Moda de ventas:", ventasModa)
ventasRango = np.ptp(ventasUnidades)
print("Rango de ventas:", ventasRango)
ventasDesviacionEstandar = np.std(ventasUnidades)
print("Desviación estándar de ventas:", ventasDesviacionEstandar)
print("************************************")
print("Estadistica Descriptiva de Gastos en Publicidad")
gastosMedia = np.mean(gastosPublicidad)
print("Media de gastos en publicidad:", gastosMedia)
gastosMediana = np.median(gastosPublicidad)
print("Mediana de gastos en publicidad:", gastosMediana)
gastosModa = stats.mode(gastosPublicidad)
print("Moda de gastos en publicidad:", gastosModa)
gastosRango = np.ptp(gastosPublicidad)
print("Rango de gastos en publicidad:", gastosRango)
gastosDesviacionEstandar = np.std(gastosPublicidad)
print("Desviación estándar de gastos en publicidad:", gastosDesviacionEstandar)
```

```python
# Calcula la correlación entre ventas y gastos en publicidad. ¿Qué tipo de relación parece existir entre ambos?
# Visualizaciones:

# Realiza un gráfico de dispersión (scatter plot) para ver la relación entre las ventas y los gastos en publicidad.
graficoDispersion = plt.scatter(gastosPublicidad, ventasUnidades, cmap='viridis', edgecolors='black', alpha=0.75)
plt.xlabel("Gastos en publicidad")
plt.ylabel("Ventas")
plt.title("Relación entre ventas y gastos en publicidad")
plt.grid(True, linestyle='--', alpha=0.5)
sns.set(style="darkgrid")
plt.show()
# Haz un histograma de las ventas y de los gastos en publicidad para visualizar su distribución.
graficoHistograma = plt.hist(ventasUnidades, bins=10, alpha=0.5, label='Ventas', color='blue', edgecolor='black')
plt.hist(gastosPublicidad, bins=10, alpha=0.5, label='Gastos en publicidad', color='green', edgecolor='black')
plt.xlabel("Monto")
plt.ylabel("Frecuencia")
plt.title("Distribución de ventas y gastos en publicidad")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
# Haz un boxplot de las ventas y de los gastos en publicidad para observar la dispersión y los posibles valores atípicos.
graficoBoxplot = plt.boxplot([ventasUnidades, gastosPublicidad], labels=['Ventas', 'Gastos en publicidad'])
plt.ylabel("Monto")
plt.title("Distribución de ventas y gastos en publicidad")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
